# .apply() method

```python
colname = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','country','salary']
df = pd.read_csv('adult.data.csv',header=None, names = colname)
```

```python
df.head()
```

```python
df['salary'].unique()
```

```python
def encodedsalary(salary):
    if salary == ' <=50K':
        return '1'
    else:
        return '2'
```

```python
df['encoded salary'] = df['salary'].apply(encodedsalary)
```

```python
df.head()
```

# .apply() on multiple column

```python
colname = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','country','salary']
df = pd.read_csv('adult.data.csv',header=None, names = colname)
```

```python
df.head()
```

```python
def desc(row):
    age = row[0]
    sex = row[9]
    salary = row[14]
    temp = sex + ' earning'+ salary + '$'
    return temp
```

```python
df['description'] = df.apply(desc , axis='columns')
```

```python
df.head()
```
